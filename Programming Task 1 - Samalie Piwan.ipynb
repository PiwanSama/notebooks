```python
"""
    This application prompts the user to input two numbers and returns their sum, difference, product, and quotient in a tuple.
    It also determines whether an arithmetic response is even or odd
    The user can exit the application by entering 'quit' or 'exit' when prompted for input

    Attributes
    -----------
        show_menu : boolean
            A flag that determines whether the menu should be displayed
    Methods
    -----------
        perform_arithmetic(a,b)
            Takes two integers a and b, and returns tuple that contains their sum, difference, product, and quotient
        odd_even(num):
            Takes an integer num and determines whether it is an even or odd number
          
"""

#Set the flag to show the menu as false
show_menu = False

   
def perform_arithmetic(a,b):
    """
    This function takes two numbers as input from the user and returns their sum, difference, product, 
    and quotient in a tuple.
    It then prints whether each arithmetic calculation result is even or odd
    
    Parameters
    -----------
        a : int
            The first number entered by the user to be computed
        b : int
            The second number entered by the user to be computed
    Return
    -----------
        arithmetic_response : tuple
            A tuple that contains the sum, difference, product, and quotient of the two numbers entered
    """
    
    #parse the two input variables into strings
    a = int(a)
    b = int(b)
    #Compute the sum of the two input numbers
    summation = a + b
    #Compute the difference of the two input numbers
    difference = a - b
    #Compute the product of the two input numbers
    product = a * b
    #Compute the quotient of the two input numbers
    quotient = a // b
    
    #Create a tuple arithmetic_response to hold the sum, difference, product, and quotient
    arithmetic_response = (summation, difference, product, quotient)
    #Return the tuple
    return (arithmetic_response)

def odd_even(num):
    """
    This function determines whether a number is an even or odd number
    
    Parameters
    -----------
        num : int
            The number to check if is even or odd
    """
    
    #If the modulus of the num parameter and 2 is 0, the number is even.
    #Else the number is odd
    if num%2 == 0:
        print("Even")
    else:
        print("Odd")

#Use the show_menu boolean to keep the menu displayed in a loop until the user enters 'exit' or 'quit'
while(not show_menu):
    #Prompt the user for the first input number
    input_one = input("Enter two numbers to perform arithmetic, or enter 'exit' or 'quit' to close the application: \n\nFirst number: ")
    
    #Check if the user has opted to exit the application
    if(input_one=='quit'or input_one=='exit'):
        print("Closing the application...")
        #Set the show_menu flag to true to stop the while loop
        show_menu = True
    else:
        #Prompt the user for the second input number
        input_two = input("Second number: ")
        #Check if the user has opted to exit the application
        if(input_two=='quit'or input_two=='exit'):
            print("Closing the application...")
            show_menu = True
        else:
            #Call the perform_arithmetic() function and pass the two input numbers as parameters
            output_tuple = perform_arithmetic(input_one, input_two)
            print(output_tuple)
            for num in output_tuple:
                odd_even(num)
```

### Exercise 4a

```python
def calculate_average(grades):
    """
    This function takes a list of grades and returns their average
    
    Parameters
    -----------
        grades : list
            The list of grades to find the average of
    Return
    -----------
        average : integer
            The average of the grades from the grades list
    """
    
    #Initialize the average as 0
    ta_average = 0
    #Calculate the average using the sum of grades in ta_grades_list and length of ta_grades_list
    #Rounded off to 0 decimal points
    if(len(grades)>0):
        ta_average = round(sum(grades) / len(grades))
    else:
        ta_average = 0
    #Return the average
    return ta_average
```

```python
#Initialize the prompt_user variable for the sentinel-controlled while loop 
prompt_user = 0
#Create the list ta_grades_list to hold the grades entered by the user
ta_grades_list = []
#Initialize a variable to keep count of number of entries
num_students = 0

#Use the prompt_user flag to keep the menu displayed in a loop until the user enters '-1'
while(prompt_user != -1):
    if(num_students < 10):
        #Prompt the user for the student_grade
        student_grade = input("Enter student grades or enter '-1' to find the average and close the application: \nStudent Grade: ")
        #Parse the user input into integer format
        student_grade = int(student_grade)
        #Check if the user has opted to exit the application
        if(student_grade == -1):
            #Update the prompt_user flag to stop the while loop
            prompt_user = -1
        elif(student_grade > 100):
            #Prompt the user to enter a valid input
            print("Invalid input, enter a figure greater than 0 and less than 100")
        else:
            #Add the grade entered to ta_grades_list
            ta_grades_list.append(student_grade)
            #Increment the number of students
            num_students+=1
    else:
        #Break the loop once 10 entries are submitted
        break

average_grade = calculate_average(ta_grades_list)
print(f"\nThe average of the grades input is {average_grade}")
print("\nExiting the application...")
```

```python
#Import the python random library to generate random numbers
import random

#Generate a random number between 0 and 20
generated_guess = random.randint(0,20)
#Initialize the number of guesses to 0
try_count = 0
#Initialize the prompt_user variable for the sentinel-controlled while loop 
prompt_input = 0

#Use the prompt_user flag to keep the menu displayed in a loop until the user guesses the correct random number
while(prompt_input != -1):  
    input_guess = input("Guess the generated random number between 0 and 20:\n")
    input_guess = int(input_guess)
    
    #For each input, update the try count and inform the user if their input is correct 
    #or higher or lower than the random number generated
    if(input_guess > 20 or input_guess < 0):
        try_count+=1
        print(f"Your guess {input_guess} is out of range!\nYou have guessed {try_count} times so far\n")
    elif(input_guess == generated_guess):
        try_count+=1
        #Update the prompt_user flag to stop the while loop
        print(f"Correct! The random number is {generated_guess}. \nYou guessed {try_count} times in total\n")
        prompt_input = -1
    elif(input_guess < generated_guess):
        try_count+=1
        print(f"Your guess {input_guess} is lower than the random number, try again. \nYou have tried {try_count} guesses so far\n")
    elif(input_guess > generated_guess):
        try_count+=1
        print(f"Your guess {input_guess} is greater than the random number, try again \nYou have tried {try_count} guesses so far\n")
    else:
        print("Invalid input")
```
